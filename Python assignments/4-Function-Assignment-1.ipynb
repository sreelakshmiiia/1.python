```python
# Create a function that checks whether the given number is Odd or Even
class OddEven():
    def oddEven(num):
        if(num%2==0):
            print(num," is Even number")
        else:
            print(num," is Odd number")

number=int(input("Enter the number:"))
OddEven.oddEven(number)
            
```

```python
# Create a function that tells elegibility of marriage for male and female according to thei
#r age limit like 21 for male and 18 for female
class ElegiblityForMarriage():
    def Elegible(gender,age):
        if(gender.lower()=="male"):
            if(age>=21):
                print("ELIGIBLE")
            else:
                print("NOT ELIGIBLE")
        elif(gender.lower()=="female"):
            if(age>=18):
                print("ELIGIBLE")
            else:
                print("NOT ELIGIBLE")
        else:
            print("not applicable")
            
gender=input("Enter your gender:")
age=int(input("Enter your age:"))
ElegiblityForMarriage.Elegible(gender,age)
```

```python
# calculate the percentage of your 10th mark
class FindPercent():
    def percentage(num):
        total=0
        for i in range(num):
            marks=int(input(f"Subject{i+1}:"))
            total=total+marks
        print("Total=",total)
        percentage=total/(num*100)*100
        return percentage
        
numOfSubjects=int(input("Number of subjects:"))
print("Percentage:",FindPercent.percentage(numOfSubjects))
```

```python
#print area and perimeter of triangle using class and functions
class Triangle():
    def area():
        h=int(input("Height:"))
        b=int(input("Breadth:"))
        print("Area formula: (Height*Breadth)/2")
        area=h*b/2
        return area
    def perimeter():
        h1=int(input("Height1:"))
        h2=int(input("Height2:"))
        b=int(input("Breadth:"))
        print("Perimeter formula: Height1+Height2+Breadth")
        perimeter=h1+h2+b
        return perimeter
        
area=Triangle.area()
print("Area of triangle=",area)
perimeter=Triangle.perimeter()
print("Perimeter of triangle=",perimeter)
```
